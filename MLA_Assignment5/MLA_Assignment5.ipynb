```python
#importing required libraries
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading the dataset
data=pd.read_csv("kmeans-data.csv")
data.head()
```

```python
#plotting the data
plt.xlabel('x')
plt.xlabel('y')
plt.scatter(data['x'], data['y'])
plt.show()
```

```python
#finding optimum number of clusters using the elbow method
error=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(data[['x','y']])
    error.append(kmeans.inertia_)
```

```python
#plotting the error graph
plt.title('The Elbow method graph')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared error')
plt.plot(range(1,10), error)
plt.show()
```

```python
#creating and fitting the model
model=KMeans(n_clusters=3)
predicted=model.fit_predict(data)
```

```python
#plotting the data
c1=data[data.cluster==0]
c2=data[data.cluster==1]
c3=data[data.cluster==2]
plt.scatter(c1.x, c1['y'], s=100, color='black', label='cluster1')
plt.scatter(c2.x, c2['y'], s=100, color='blue', label='cluster2')
plt.scatter(c3.x, c3['y'], s=100, color='orange', label='cluster3')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1], s=200, color='yellow',label='centroid')
plt.legend()
plt.show()
```
